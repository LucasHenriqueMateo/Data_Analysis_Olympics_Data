##**Limpeza de dados**

```python
#Percebemos valores nulos em população e PIB per capita
#Dando uma olhada nestes valores
dic[dic.isnull().any(axis = 1)]
```

```python
#Concatenando os DataFrames e arrumando o index
olimpiadas = pd.concat([verao,inverno], axis = 0, keys = ['Summer', 'Winter'],
          names = ['Edition']).reset_index().drop(columns = 'level_1')
```

```python
olimpiadas.head(3)
```

```python
dic.head(3)
```

```python
#Aplicando o join dos dados da olimpiadas com o dado do dicionário
#Obtendo a coluna do país por escrito e excluindo a coluna de país com siglas.
olimpiadas = olimpiadas.merge(dic.iloc[:, :2], how = 'left',
               left_on = 'Country', right_on = 'Code').drop(columns = ['Code'])
olimpiadas            
```

```python
#Renomeando as colunas
olimpiadas.rename(columns = {'Country_x': 'Code', 'Country_y': 'Country'}, inplace = True)
```

```python
dic
```

```python
#Tirando o espaço do nome da coluna
dic.rename(columns = {'GDP per Capita':'GDP'}, inplace = True)
dic
```

```python
#Verificando valores faltantes(missing values)
olimpiadas.info()
```

```python
#Detectamos a coluna de País com 6k de NaN
olimpiadas.loc[olimpiadas['Country'].isnull()]
```

```python
#Guardando em uma variável linhas com valores para países ausentes
missing_country = olimpiadas.loc[olimpiadas.Country.isnull()].index
missing_country
```

```python
#Verificada a linha dos países nulos e retira o código do país
#Pela análise, nota-se que tratam-se de países que não existem mais
olimpiadas.loc[olimpiadas.Country.isnull()].Code.value_counts()
```

```python
#Guardando países que não existem mais em uma variável
old_countries = olimpiadas.loc[olimpiadas.Country.isnull(), :].Code.value_counts().index
old_countries
```

```python
#Com uma breve pesquisa no google é possível identificar quais são os países.
#Renomeando os países com mapper
mapper = pd.Series(index=old_countries, name = "Country", data = ["Soviet Union", "East Germany", "Romania", "West Germany", "Czechoslovakia",
                               "Yugoslavia", "Unified Team", "Unified Team of Germany", "Mixed teams", "Serbia",
                              "Australasia", "Russian Empire", "Montenegro", "Trinidad and Tobago", "Bohemia", 
                              "West Indies Federation", "Singapore", "Independent Olympic Participants"])
mapper
```

```python
#Localizando os valores faltantes e aplicando o map
olimpiadas.loc[missing_country, 'Code'].map(mapper)
```

```python
#Preenchendo os valores faltantes com os devidos países
olimpiadas['Country'].fillna(olimpiadas['Code'].map(mapper), inplace = True)
```

```python
olimpiadas.loc[missing_country]
```

##**Testes de hipóteses**

#Hipótese 1 - Não há relação entre o total de medalhas e a população do país

```python
import scipy.stats as stats
```

```python
r, p_value = stats.spearmanr(medalhas_por_pais['Total'], medalhas_por_pais['Population'], 
                             nan_policy = "omit")
```

```python
#Nota-se um valor muito baixo de p-value, portanto podemos rejeitar a hipótese 1.
r.round(2), p_value.round(3)
```

***Podemos concluir que existe uma relação significativa(+) entre o total de medalhas e a população.***

#Hipótese 2 - Não há relação entre o total de medalhas e o PIB do país

```python
r, p_value = stats.spearmanr(medalhas_por_pais.Total, medalhas_por_pais.GDP,
                             nan_policy = "omit")
```

```python
#Nota-se um valor muito baixo de p-value, portanto podemos rejeitar a hipótese 2.
r.round(2), p_value.round(3)
```

***Podemos concluir que existe uma relação significativa(+) entre o total de medalhas e o PIB per capita.***

#Hipótese 3 - Não existe relação entre o total de medalhas e participações em olimpíadas

```python
medalhas_por_pais.head(10)
```

```python
r, p_value = stats.spearmanr(medalhas_por_pais.head(50).Total, 
                             medalhas_por_pais.head(50).Total_Games)
```

```python
#Nota-se um valor muito baixo de p-value, portanto podemos rejeitar a hipótese 2.
r.round(2), p_value.round(3)
```

***Podemos concluir que existe uma relação significativa(+) entre o total de medalhas e participações.***

#**Ranking**

```python
#Crosstab recebendo os dados da edição e país
verao_x_inverno = pd.crosstab(olimpiadas['Country'], olimpiadas['Edition'])
verao_x_inverno
```

```python
#Crosstab recbendo os dados do gênero e país
homem_x_mulher =  pd.crosstab(olimpiadas['Country'], olimpiadas['Gender'])
homem_x_mulher
```

```python
#Concatenando as tabelas
medalha_por_categoria = pd.concat([verao_x_inverno, homem_x_mulher], axis = 1)
medalha_por_categoria
```

```python
medalha_por_categoria["Total"] = medalha_por_categoria.Summer + medalha_por_categoria.Winter
medalha_por_categoria
```

```python
medalha_por_categoria.sort_values("Total", ascending = False, inplace = True)
medalha_por_categoria
```

```python
#Ranking de medalhas por categoria utilizando a média
ranks = medalha_por_categoria.rank(ascending = False, method = "average")
ranks
```

```python
#Top 30 - Por estação, gênero e total
top_30 = ranks.head(30)
top_30
```

```python
#Mapa de calor
plt.figure(figsize = (50, 10))
sns.heatmap(top_30.T, cmap='RdYlGn_r', annot=True, fmt='2.0f')
plt.show()
```

***Podemos tirar diversas conclusões, como no Reino Unido, os homens têm melhor desempenho comparado as mulheres.***


#**Jogos de verão x Jogos de inverno**

```python
top_30.sort_values("Summer")
```

```python
#Mapa de calor para os jogos de verão
plt.figure(figsize = (50, 10))
sns.heatmap(top_30.sort_values(by = "Summer").T, cmap='RdYlGn_r', annot=True, fmt='2.0f')
plt.show()
```

***Nota-se que o Brasil e Cuba têm uma forte relação com os jogos de verão, o que faz sentido levando em consideração o clima dos países.***

```python
top_30.sort_values("Winter")
```

```python
#Mapa de calor para os jogos de inverno
plt.figure(figsize = (50, 10))
sns.heatmap(top_30.sort_values(by = "Winter").T,cmap='RdYlGn_r',annot=True,fmt='2.0f')
plt.show()
```

***Nota-se que países como Canadá, Noruega, Austria e República Theca tem melhores desempenhos em jogos de Inverno, o que faz sentido, levando em consideração que são países com invernos longos e climas não muito quentes no verão.***




```python
#Subtraindo as medalhas de verão por medalhas de inverno
rank_diff = top_30['Summer'].sub(top_30['Winter']).sort_values().to_frame()                                     
rank_diff
```

```python
#Mapa de calor de diferença de medalhas por estação
plt.figure(figsize = (35, 5))
sns.heatmap(rank_diff.T, cmap='RdBu', annot=True, fmt='2.0f', center = 0)
plt.xlabel("País", fontsize = 18)
plt.show()
```

Vemos que o Brasil é o país com a maior diferença entre número de medalhas em olimpíadas de verão, e número de medalhas em olimpíadas de invero.
Por outro lado, a Áustria representa um desempenho melhor em olimpíadas de verão.
